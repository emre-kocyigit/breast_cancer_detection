## Supervised Learning

When inputs and their expected outputs are given to ML model during the training.

- Expected output = Label = Target

To apply supervised learning, you should have labelled data.

### 0. Import "three musketeers'" of the ML world!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### 1. Load the data

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df['label'] = data.target
```

### 4. Split the data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('label', axis=1)
```

```python
y = df['label']
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### 5. Feature improvement

```python
# normalization
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

### 6. Build Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(X_train_scaled, y_train)
```

```python
y_pred = log_model.predict(X_test_scaled)
```

### 7. Evaluate the model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
# roc auc
from sklearn.metrics import roc_auc_score
```

```python
y_probs = log_model.predict_proba(X_test_scaled)[:, 1]
```

```python
print("ROC AUC score:", roc_auc_score(y_test, y_probs))
```

### 8. Compare the scalers

```python
# define scalers
scalers = [('StandardScaler', StandardScaler()),
           ('MinMaxScaler', MinMaxScaler()),
           ('RobustScaler', RobustScaler())]
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for scaler in scalers:
    pipeline = Pipeline([scaler,
                         ('logistic_regression', LogisticRegression())])
    scores = cross_val_score(pipeline, X, y, cv=5)
    print(f"{scaler[0]}: Mean Accuracy: {scores.mean():.4f}, Std Dev: {scores.std():.4f}")
```

### 9. Hyper-parameter Optimization

```python
from sklearn.model_selection import GridSearchCV
```

```python
# hyperparameters
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs', 'newton-cholesky','saga']
}
```

```python
model = LogisticRegression(max_iter=1000)
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
best_model = grid_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test_scaled)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```
